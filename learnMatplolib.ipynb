### What is Matplotlib
[Matplotlib](https://matplotlib.org/) is the most popular library for creating static, animated, and "interactive" visualizations in Python (began in 2003).     

Produce static images (png default) and predates HTML5. Inspiration from Matlab.

````
# Let's have graphs side by side
# data
x=[1,2,3,4]
y1=[3,8,23,44]
y2=[6,12,12,48]
y3=[10,20,33,54]
````

````
#@title
plt.style.use("fivethirtyeight")
plt.figure(figsize=(10,5))

plt.suptitle("Three graph")

plt.subplot(1,3,1) # nb rows, nb cols, index
plt.plot(x,y1,'p-',linewidth=2, markersize=12, label="apples")

plt.subplot(1,3,2)
plt.plot(x,y2,'o--',linewidth=4, markersize=1, label="oranges",color="orange")
plt.xlabel("Time")
plt.ylabel("Numbers")
plt.title("Orange")

plt.subplot(1,3,3)
plt.plot(x,y3,'D:',linewidth=1, markersize=8,label="potatoes", color="grey")
plt.xlabel("Time")
plt.ylabel("Numbers")
plt.title("Potatoes");
````

Same thing but with the object oriented way.

````
#@title
fig, ax = plt.subplots(1,3) # another way : fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(10,5))

fig.set_size_inches((10,5))

ax[0].plot(x,y1, 'p-')
ax[1].plot(x,y2, "o--", color="orange")
ax[2].plot(x,y3, 'D:', color="grey")

ax[0].set_xlabel("Time")
ax[0].set_ylabel("Number")
ax[0].set_title("apples")

ax[1].set_xlabel("Time")
ax[1].set_title("oranges")

ax[2].set_xlabel("Time")
ax[2].set_title("potatoes");
````

### Pie charts <a name="pieCharts"></a>

Works well with few data, let's take the population data by countries.

````
# data
countries=["Italy","France","UK","Germany"]
pop=[59.2,65.4,66.9,82.4]
````

````
#@title
plt.figure(figsize=(5,5))
plt.pie(pop, labels=countries,shadow=True);
````

````
#@title
plt.figure(figsize=(5,5))
plt.pie(pop, labels=countries,shadow=True,explode=[0,0.1,0,0],startangle=180);
````

````
#@title
plt.figure(figsize=(5,5))
wedgeprops={"edgecolor":"k",'linewidth': 2, 'linestyle': 'dashed', 'antialiased': True}
plt.pie(pop, labels=countries,shadow=True,explode=[0,0.1,0,0],startangle=180,
        wedgeprops=wedgeprops, autopct='%.2f%%');
````

### Bar charts (simple, multi) <a name="barCharts"></a>

````
# create a bar graph with countries populations
# data
countries=["Italy","France","UK","Germany"]
pop=[59.2,65.4,66.9,82.4]
````

````
#@title
plt.bar(countries,pop)
plt.savefig("francePop.svg")
````

````
# the same kind of bar graph but horizontal
# data
countries=["Italy","France","UK","Germany"]
pop=[59.2,65.4,66.9,82.4]
color=["green","blue","red","yellow"]
variance=[1,2,3,4] # error margin, obvioulsy not true
````

````
#@title
plt.barh(countries, pop, color=color, edgecolor="black",xerr=variance)
plt.title("Populations")
plt.xlabel("Millions of people")
plt.ylabel("Countries")
plt.savefig("francePop.svg")
````

Same but oject oriented,

````
#@title
fig, ax=plt.subplots()
ax.barh(countries, pop, xerr=variance, color=color, edgecolor="black",)
ax.set_title("Populations")
ax.set_xlabel("Millions of people")
ax.set_ylabel("Countries")
# adavanced : writing values inside
for i, v in enumerate(pop):
  ax.text(pop[i]+5, i, pop[i])
````

````
# Two bar graphs comparison
# data
divisions=["Div1","Div2","Div3","Div4","Div5"]
girls=[16,11,18,7,14]
boys=[17,10,15,10,12]
````

````
#@title
index= np.arange(5)
barWidth=0.3
plt.bar(index,girls, width=0.3,label="Girls")
plt.bar(index+0.3,boys, width=0.3,label="Boys")
plt.title("Comparing 2 categories")
plt.xlabel("Divisions")
plt.ylabel("Boys vs average")
plt.xticks(index+barWidth/2,divisions)
plt.legend(loc="best");
````

````
#@title
index= np.arange(5)
barWidth=0.3
plt.bar(index,girls, width=0.3,label="Girls")
plt.bar(index,boys, width=0.3,label="Boys",bottom=girls)
plt.xticks(index,divisions)
plt.title("Comparing 2 categories")
plt.xlabel("Divisions")
plt.ylabel("Boys vs girls")
plt.legend(loc="best");
````

### Stack plots <a name="stackPlots"></a>

````
# data from United Nations World Population Prospects (Revision 2019)
# data https://population.un.org/wpp/, license: CC BY 3.0 IGO
year = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2018]
continents=["africa","americas","asia","europe","oceania"]
popAfrica=[228, 284, 365, 477, 631, 814, 1044, 1275]
popAmericas=[340, 425, 519, 619, 727, 840, 943, 1006]
popAsia= [1394, 1686, 2120, 2625, 3202, 3714, 4169, 4560]
popEurope=[220, 253, 276, 295, 310, 303, 294, 293]
popOceania=[12, 15, 19, 22, 26, 31, 36, 39]
````

````
#@title
plt.figure(figsize=(10,5))
plt.stackplot(year,
              [popAfrica, popAmericas,popAsia,popEurope,popOceania],
              labels=continents);
plt.legend(loc="upper left");
````

````
# same from matplotlib examples but object notation and python dictionary usage
# data
year = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2018]
population_by_continent = {
    'africa': [228, 284, 365, 477, 631, 814, 1044, 1275],
    'americas': [340, 425, 519, 619, 727, 840, 943, 1006],
    'asia': [1394, 1686, 2120, 2625, 3202, 3714, 4169, 4560],
    'europe': [220, 253, 276, 295, 310, 303, 294, 293],
    'oceania': [12, 15, 19, 22, 26, 31, 36, 39],
}

#@title
fig, ax = plt.subplots()
ax.stackplot(year, population_by_continent.values(),
             labels=population_by_continent.keys())
ax.legend(loc='upper left')
ax.set_title('World population')
ax.set_xlabel('Year')
ax.set_ylabel('Millions of people)')
fig.savefig("worldPop.svg")
````

### 3D Plots <a name="3Dplots"></a>

````
# data
pos=np.random.rand(3,100) # 3 series of 100 random numbers between 0 and 1





# code
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter3D(xs=pos[0],ys=pos[1],zs=pos[2])
````

````
# you can sets limits, multiply a numpy array...
# data
pos=np.random.rand(3,100)
# code
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.set_xlim(0,10)
ax.set_ylim(0,10)
ax.set_zlim(0,10)
ax.scatter3D(xs=pos[0]*10,ys=pos[1]*10,zs=pos[2]*10)
````

````
# plotting an equqtion in 3D 
# data
z_line = np.linspace(0, 15, 1000)
x_line = np.cos(z_line)
y_line = np.sin(z_line)





# code
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
plt.xticks(np.arange(-1, 1, step=0.5)) # optional : improving axis reading
ax.plot3D(x_line, y_line, z_line, 'green')
````
